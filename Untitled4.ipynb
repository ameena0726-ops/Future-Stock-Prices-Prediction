```python
import yfinance as yf
import pandas as pd

stock= yf.download ("AAPL", start="2020-01-01", end="2024-01-01")
print (stock.head())
```

```python
#Features and Targets
X= stock[['Open','High','Low','Volume']]
Y= stock['Close']
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train ,X_test ,Y_train, Y_test = train_test_split(X,Y,test_size=0.2,shuffle=False)
```

```python
#Model Training
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(X_train, Y_train)
```

```python

#Prediction
predictions = model.predict(X_test)
```

```python
#Graph
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(Y_test.values, label="Actual Price",color="red")
plt.plot(predictions, label="Predicted Price",color="green")

plt.legend()
plt.title("Actual vs Predicted Stock Prices")
plt.show()
```

```python
#Next Day Prediction
next_day = model.predict([X.tail(1).values[0]])
print("Next Day Predicted Close Price:", next_day)
```
